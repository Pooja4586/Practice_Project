To convert your GPA to a 4.0 scale:

Letter Grade Percent Grade 4.0 Scale 
A+ 97-100 4.0 
A 93-96 4.0 
A- 90-92 3.7 
B+ 87-89 3.3 
B 83-86 3.0 
B- 80-82 2.7 
C+ 77-79 2.3 
C 73-76 2.0 
C- 70-72 1.7 
D+ 67-69 1.3 
D 65-66 1.0 
E/F Below 65 0.0

```python
for column in data.columns:

    data[column]=data[column].replace('A+',4.0)
    data[column]=data[column].replace('A',4.0)
    data[column]=data[column].replace('A-',3.7)
    data[column]=data[column].replace('B+',3.4)
    data[column]=data[column].replace('B',3.0)
    data[column]=data[column].replace('B-',2.7)
    data[column]=data[column].replace('C+',2.4)
    data[column]=data[column].replace('C',2.0)
    data[column]=data[column].replace('C-',1.7)
    data[column]=data[column].replace('D+',1.4)
    data[column]=data[column].replace('D',1.0)
    data[column]=data[column].replace('F',0.0)
```

Model Building and Preprocessing

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop(['CGPA'],axis=1)
Y=data['CGPA']
```

```python
#Data Tranformation
from sklearn.preprocessing import power_transform
X=power_transform(X,method='yeo-johnson')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score,mean_squared_error
from sklearn import linear_model

from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=20)
```

```python
#Linear Regression

lr = LinearRegression()
lr.fit(X_train, y_train)

Y_pred = lr.predict(X_train)#training prediction
Y_pred1 = lr.predict(X_test)#test prediction
r2_lr=r2_score(y_test,Y_pred1)
print("Accuracy Training: ",r2_score(y_train,Y_pred))
print("Accuracy Test: ",r2_lr)
```

```python
from sklearn.model_selection import cross_val_score

for j in range(2,10):
    cscore_lr=cross_val_score(lr,X,Y,cv=j)
    cmean_lr=cscore_lr.mean()
print("Cross Value Scores: ", cscore_lr)
print("\nCross Value Mean ",cmean_lr)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

parameters={'alpha':[.0001,.001,.01,.1,1,10],
           'random_state':(range(0,10))}

ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(X_train,y_train)

print(clf.best_params_)
```

```python
#lasso Fitting

ls=Lasso(alpha=0.001, random_state=0)
ls.fit(X_train,y_train)
ls_score=ls.score(X_train,y_train)

print( mean_squared_error(ls.predict(X_train),y_train))
ls_score*100
```

```python
lspred=ls.predict(X_test)
r2_lss=r2_score(y_test,lspred)
print(r2_lss)
```

```python
for j in range(2,10):
    ls_cscore=cross_val_score(ls,X,Y,cv=j)
    ls_cmean=ls_cscore.mean()
print("Cross Value Scores: ", ls_cscore)
print("\nCross Value Mean ",ls_cmean)
```

```python
#Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor()
parameters={'criterion':['mse','mae'],'max_features':['auto','sqrt','log2']}

clf1=GridSearchCV(rfr,parameters)
clf1.fit(X_train,y_train)

print(clf1.best_params_)

```

```python
rfr=RandomForestRegressor(criterion= 'mae', max_features='log2')

rfr.fit(X_train,y_train)
rfr_score=rfr.score(X_train,y_train)

rfr_pred=rfr.predict(X_test)
r2_rfr=r2_score(y_test,rfr_pred)
print(r2_rfr)
```

```python
for j in range(2,10):
    rfr_cscore=cross_val_score(rfr,X,Y,cv=j)
    rfr_cmean=rfr_cscore.mean()
print("Cross Value Scores: ", rfr_cscore)
print("\nCross Value Mean ",rfr_cmean)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr=GradientBoostingRegressor()

parameters={'loss':['squared_error', 'absolute_error', 'huber', 'quantile'],'criterion':['friedman_mse', 'squared_error']}

clf2=GridSearchCV(gbr,parameters)
clf2.fit(X_train,y_train)

print(clf2.best_params_)
```

```python
#Gradient Boosting Regressor

gbr=GradientBoostingRegressor(criterion= 'squared_error', loss='squared_error')
gbr.fit(X_train,y_train)
gbr_score=gbr.score(X_train,y_train)

gbr_pred=gbr.predict(X_test)

r2_gbr=r2_score(y_test,gbr_pred)

print(r2_gbr)
```

```python
for j in range(2,10):
    gbc_cscore=cross_val_score(gbr,X,Y,cv=j)
    gbc_cmean=gbc_cscore.mean()
print("Cross Value Scores: ", gbc_cscore)
print("\nCross Value Mean ",gbc_cmean)
```

```python
#Decision Tree Regressor

from sklearn.tree import DecisionTreeRegressor

dtr=DecisionTreeRegressor()
parameters={'criterion':['squared_error','friedman_mse','poisson','absolute_error']}

clf3=GridSearchCV(dtr,parameters)
clf3.fit(X_train,y_train)

print(clf3.best_params_)

```

```python
dtr=DecisionTreeRegressor(criterion= 'friedman_mse')
dtr.fit(X_train,y_train)
dtr_score=dtr.score(X_train,y_train)

dtr_pred=dtr.predict(X_test)

r2_dtr=r2_score(y_test,dtr_pred)

print(r2_dtr)
```

```python
for j in range(2,10):
    dtr_cscore=cross_val_score(dtr,X,Y,cv=j)
    dtr_cmean=gbc_cscore.mean()
print("Cross Value Scores: ", dtr_cscore)
print("\nCross Value Mean ",dtr_cmean)
```
